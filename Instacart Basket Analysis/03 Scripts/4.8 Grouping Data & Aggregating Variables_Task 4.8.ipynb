# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
## Create “max_order” column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create a flag using loc - create Loyal customer max_order condition

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create a flag using loc - create Regular customer max_order condition

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create a flag using loc - create 'New customer' max_order condition

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequency

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check outcome

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# 5. Spending habits of the three types of customers

```python
# Split the data into groups based on “loyalty_flag and apply the agg() function to obtain the mean, min and max values

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Loyal customers have the lowest average price while new customers have the highest. 

# 6. . Create a spending flag for each user based on the average price

```python
## Create “mean_order” column

ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create flag to assign "spending" labels to user ID based on mean product price

ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequency

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
## Create “median_days” column

ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Create flag to assign frequency labels to user ID based on mean product price

ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Check null values

print(ords_prods_merge[ords_prods_merge['frequency_flag'].isnull()==True])
```

### The null values are because this customer only had placed one order and the median day can't be calculated. This customer can be put under the non-frequent customer category

```python
# Replace missing values

ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
# Check frequency

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
